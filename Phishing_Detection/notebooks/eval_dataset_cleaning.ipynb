```python
import pandas as pd

# Load the new dataset
eval_df = pd.read_csv("/content/drive/MyDrive/PhiUSIIL_Phishing_URL_Dataset.csv")

# Show structure
print("Dataset shape:", eval_df.shape)
print("Columns:", eval_df.columns)
eval_df.head()
```

```python
eval_df = eval_df[["URL", "label"]].copy()

# Rename for consistency (optional but good practice)
eval_df.rename(columns={"URL": "url"}, inplace=True)

# Lowercase the URLs
eval_df["url"] = eval_df["url"].str.lower()

# Remove http:// or https://
eval_df["url"] = eval_df["url"].str.replace(r"^https?:\/\/", "", regex=True)

# Remove trailing slashes
eval_df["url"] = eval_df["url"].str.rstrip("/")

# Drop rows with missing values
eval_df.dropna(subset=["url", "label"], inplace=True)

# Show final shape and class balance
print("Final shape:", eval_df.shape)
print(eval_df["label"].value_counts())
eval_df.head()
```

```python
# Save preprocessed evaluation dataset to Drive
eval_df.to_csv("/content/drive/MyDrive/PhiUSIIL_Phishing_URL_Eval_Preprocessed.csv", index=False)

print("✅ Preprocessed dataset saved successfully!")
```
